## 1. La función `plot`

Nota: Recuerde siempre revisar la documentación de las funciones.

Comencemos definiendo un conjunto de datos con los que trabajaremos.

```r
data
```

Comencemos visualizando las variable1 y variable2 mediante los puntos representados en un plano.

```r
plot(x = data$variable1, y = data$variable2)
```

A continación, veamos distintas opciones relacionadas con la manipulación de gráficos que nos ayudarán a mejorar las presentaciones de los mismos. 

**title**

`title` permite indicar el título de un gráfico, el mismo debe indicarse como el llamado a la función, a continuación del llamado a la función `plot`

```r
# Agregar título al gráfico
plot(data$variable2, data$variable1)

title("Dispersión de las variable1 y variable2")
```

## 2. Parámetros `par`

La función `par` permite modificar varios parámetros de los gráficos. Veamos algunos ejemplos

```r
?par
```

**cex** es un parámetro numérico que permite incrementar o disminuir textos y símbolos en el gráfico

```r
# Modificar el tamaño del texto en el gráfico con 'cex'

# Aumentar el tamaño del título
par(cex.main=1.5)

# Disminuir el tamaño de las etiquetas
par(cex.lab=0.7)

# Disminuir el tamaño de los ejes
par(cex.axis=0.5)

# Generar el gráfico con el título
plot(data$variable2, data$variable1)
title("Dispersión de las variable1 y variable2")
```

Los parámetros pueden indicarse de a uno, o bien todos juntos, por ejemplo:

```r
# Modificar el tamaño del texto en el gráfico con 'cex'

# Aumentar el tamaño del título, etiquetas y ejes
par(cex.main=1.5,
    cex.lab=0.7,
    cex.axis=0.5)

# Generar el gráfico con el título
plot(data$variable2, data$variable1)
title("Dispersión de las variable1 y variable2")
```

Para obtener el listado de los valores actuales, basta con llamar a la función sin parámetros

```r
par()
```

## 3. Algunos tipos de gráficos

### Gráficos de líneas

Los gráficos de líneas pueden ser visualizados con la función plot, modificando el parámetro `type = 'l'`. Veamos un ejemplo

```r
plot(data$variable1, type = 'l')
```

Asimismo se pueden modificar varios parámetros. Por ejemplo, 

* tipo de líneas
* nombres de las variables
* color de líneas

```r
# Modificar las etiquetas
plot(data$variable1, type = 'o', 
     ylab = "variable 1", xlab = "índice",
    col = "blue")
```

**Múltiples líneas**

La función `lines` permite agregar datos a un gráfico ya generado. Veamos su uso

```r
# Agregar una nueva línea de datos 

# Generar el gráfico
plot(data$variable1, type = 'l', 
     ylab = "variables", xlab = "índice",
    col = "blue")

lines(data$variable2, type = 'l', col = "brown")
```

### Histograma

La función `hist` permite generar histogramas a partir de un conjunto de datos. Veamos un ejemplo con la *variable1* de nuestro conjunto de datos.

```r
hist(data$variable1)
```

### Boxplot

Los boxplot o diagramas de caja nos dan sobre la distribución de un conjunto de datos.

```r
boxplot(data$variable1)
```

### Gráfico de barras

```r
# Generar un gráfico de barras a partir de un conjunto de datos con 
# distribución normal
v_norm <- sample(1:5, 1000, replace = TRUE, prob = c(0.1, 0.2, 0.4, 0.2, 0.1))

barplot(table(v_norm))
```

### Etiquetas de datos

La función `text` permite agregar texto a los gráficos. Uno de los usos más frecuente es el de agregar etiquetas.

Nota: No olvide consultar la documentación mediante `?text`

Veamos un ejemplo en el que se requiere agregar etiquetas de datos en un histograma.

Inicialmente, almacenamos el histograma en una variable. Luego, hacemos uso de la función `text` indicando los parámetros correspondientes.

```r
# Generar un histograma y almacenar en variable
h <- hist(data$variable1)

# Agregar etiquetas de datos en el histogramap
text(x = h$mids, y = h$counts, labels=h$counts, adj=c(0.5, -0.5))
```

En el ejemplo, los atributos *mids* y *counts* corresponden a los puntos medios de cada celda, y al valor de cada celda respectivamente. 

Para más información consulte la documentación de la función con

`?hist`

```r
?hist
```

## Guardar gráficos en archivos

Los gráficos generados se pueden guardar como archivos de distintos formatos.

R ofrece una variedad de herramientas para este fin, incluyendo **pdf, jpeg, png** y otros. 

Para exportar un gráfico a un archivo, se requieren los siguientes pasos:

1. Definir el archivo de salida, incluyendo el nombre y formato
2. Definir el gráfico a exportar
3. Cerrar el archivo

```r
# Guardar gráfico en archivo

# 1. Definir el archivo de salida
jpeg(filename = "../images/histograma_var1.jpg")

# 2. Generar el gráfico a exportar
hist(data$variable1)

# 3. Cerrar el archivo
dev.off()
```

Así como es posible generar archivos de imágenes mediante la función `jpeg`, también es posible generar archivos con otros formatos. Algunas de estas funciones incluyen:

* `pdf`
* `png`
* `postscript`
* `bmp`
